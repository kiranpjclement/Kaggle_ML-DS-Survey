### Renaming 

```python

df['Q4']=df['Q4'].replace(['No formal education past high school'],'High_School')
df['Q4']=df['Q4'].replace(['Some college/university study without earning a bachelorâ€™s degree'],'Private_College')
df['Q4']=df['Q4'].replace(['I prefer not to answer'],'Prefer_not_answer')
df['Q11']=df['Q11'].replace(['A cloud computing platform (AWS, Azure, GCP, hosted notebooks, etc)'],'Cloud_Platform')
df['Q11']=df['Q11'].replace(['A deep learning workstation (NVIDIA GTX, LambdaLabs, etc)'],'GPU')
df['Q11']=df['Q11'].replace(['A personal computer / desktop'],'PC/Desktop')
```

## Q3 COUNTRY

```python
filt_top10=(df['Q3'] == 'India')| (df['Q3'] == 'USA')| (df['Q3'] == 'USA')| (df['Q3'] == 'Japan')| (df['Q3'] == 'China')|(df['Q3'] == 'Brazil')| (df['Q3'] == 'Russia')| (df['Q3'] == 'Nigeria')| (df['Q3'] == 'UK')| (df['Q3'] == 'Pakistan')|(df['Q3'] == 'Eygpt')
sns.set(style="whitegrid")
plt.figure(figsize=(12,9))
total = float(len(df))
ax = sns.countplot(x=df['Q3'].loc[filt_top10].loc[filt_top10],palette="Blues_d", data=df,order=df['Q3'].loc[filt_top10].value_counts().index)
plt.title('Top 10 Countries in the Survey', fontsize=20)
ax.set(xlabel=" Countries", ylabel = "Count")
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='right')
plt.show()
```
